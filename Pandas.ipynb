```python
import pandas as pd
```

```python
x = pd.Series(["Geeks", "Hello", "apples", 5, True])    # Series is a column vector
y = pd.Series([i for i in range(1, 11, 2)])       # List comprehension

dataframe = pd.DataFrame({"Col1": x, "Col2": y})

dataframe["Ratings"] = pd.Series([1, 2, 3, 4, 5])
dataframe
```
